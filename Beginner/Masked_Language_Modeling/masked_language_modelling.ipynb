```python
# Exercise: Write a function that automatically masks and replaces words
# in a whole document. You might choose which words to replace based on some
# statistic, e.g. TF-IDF.
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z']+\b")
tfidf.fit(df['text'])
idf_map = dict(zip(tfidf.get_feature_names_out(), tfidf.idf_))
```

```python
word_re = re.compile(r"[a-zA-Z][a-zA-Z']+")

def mask_and_replace_sentence(sentence, mlm, idf_map, pct=0.15, min_word_len=3):
    words = word_re.findall(sentence)
    candidates = [w for w in set(words) if len(w) >= min_word_len]
    if not candidates:
        return sentence

    scores = {w: idf_map.get(w.lower(), 0) for w in candidates}
    n_to_mask = max(1, int(len(candidates) * pct))
    top_words = sorted(candidates, key=scores.get, reverse=True)[:n_to_mask]

    mask_token = mlm.tokenizer.mask_token
    for w in top_words:
        pattern = re.compile(r'\b' + re.escape(w) + r'\b')
        masked_sentence, n_subs = pattern.subn(mask_token, sentence, count=1)
        if n_subs == 0:
            continue

        preds = mlm(masked_sentence)
        if not preds:
            continue

        replacement = preds[0]['token_str'].strip()
        sentence = masked_sentence.replace(mask_token, replacement, 1)

    return sentence
```

```python
def mask_and_replace_document(doc, mlm, idf_map, pct=0.15, min_word_len=3):
    paragraphs = doc.split('\n\n')
    new_paragraphs = []
    for para in paragraphs:
        sentences = re.split(r'(?<=[.!?])\s+', para.strip())
        new_sentences = [
            mask_and_replace_sentence(s, mlm, idf_map, pct, min_word_len)
            for s in sentences if s
        ]
        new_paragraphs.append(' '.join(new_sentences))
    return '\n\n'.join(new_paragraphs)
```

```python
new_doc = mask_and_replace_document(doc, mlm, idf_map, pct=0.15)

print("ORIGINAL:\n")
print(textwrap.fill(doc, replace_whitespace=False, fix_sentence_endings=True))
print("\n\nMODIFIED:\n")
print(textwrap.fill(new_doc, replace_whitespace=False, fix_sentence_endings=True))
```
